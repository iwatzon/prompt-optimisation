```python
from dotenv import load_dotenv
import os

load_dotenv()
hf_api_token = os.getenv("HF_API_TOKEN")
openai_api_key = os.getenv("OPENAI_API_KEY")
```

```python
from typing import List
from langchain.chat_models import ChatOpenAI
from langchain_community.llms import HuggingFaceEndpoint
from langchain.agents import AgentExecutor, create_openai_tools_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

class EvoAlgAgent:

    def __init__(self, n: int):
        self.n = n

    def create_openai_agent(self, model: str, tools: list, system_prompt: str):
        # Each worker node will be given a name and some tools.
        prompt = ChatPromptTemplate.from_messages(
            [
                (
                    "system", 
                    system_prompt,
                ),
                ("human", "{input}"),
                ("placeholder", "{agent_scratchpad}"),
            ]
        )
        llm = ChatOpenAI(
            api_key=openai_api_key,
            temperature=1.0,
            model=model
        )
        agent = create_openai_tools_agent(llm, tools, prompt)
        executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
        return executor

    def generate(self, task: str, tools: list):
        """
        Agent that executes an evolutionary algorithm to generate prompts
        """

        system_prompt = """
        You are an expert in prompt engineering and have been tasked with generating a set of prompts for a specific task.
        You must generate the prompts using an evolutionary algorithm. You have access to tools to help with each stage of the process.
        """

        instruction = f"""
        You are an agent designed to execute an evolutionary algorithm to generate effective prompts for the following task: {task}.
        The process of the evolutionary algorithm must be as follows:

        **Evolutionary Algorithm**
        1. Generate a population of {self.n} prompts. Use the thinking_styles tools to guide this. Choose thinking styles that are most relevant to the task. 
        2. Generate new prompts by combining prompts from the population that complement each other. Combine prompts based on style and content.
        3. Mutate the new prompts. Use the mutator_prompts tools to guide this. Mutate the prompts to introduce new elements, perspectives, or challenges.
        4. Evaluate the new prompts. Use the evaluate_prompts tool for this. 
        5. Rank the prompts based on their performance. Use the rank_prompts tool.
        6. Repeat steps 1-5 for 10 iterations. Step 1 should include the child prompts from the previous iteration and new prompt generation should be guided by feedback and rankings.

        Once the evolutionary algorithm is executed, provide the final set of generated prompts with their rankings.
        """

        agent = self.create_openai_agent("gpt-4o", tools, system_prompt)
        results = agent.invoke({"input": instruction})

        return results
```

```python
from langchain.tools import tool

@tool
def thinking_prompts():
    """
    Returns a list of prompts that can be used to guide thinking styles.
    """
    return thinking_styles

@tool
def mutator_prompts():
    """
    Returns a list of prompts that can be used to guide mutations.
    """
    return mutators

@tool
def evaluate_prompts(prompts: List[str]):
    """
    Evaluates a selection of prompts providing pros and cons.
    """
    chat = ChatOpenAI(model_name="gpt-4o", temperature=1.0)
    instruction = f"""Evaluate the following prompts: {prompts}. Provide pros and cons for each prompt"""
    return chat.invoke(instruction).content

@tool
def rank_prompts(prompts: List[str]):
    """
    Ranks a selection of prompts based on effectiveness.
    """
    chat = ChatOpenAI(model_name="gpt-4o", temperature=1.0)
    instruction = f"""Rank the following prompts based on effectiveness: {prompts}. Provide reasoning for each ranking"""
    return chat.invoke(instruction).content
```

```python
evo_agent = EvoAlgAgent(10)
```

```python
tools = [thinking_prompts, mutator_prompts, evaluate_prompts, rank_prompts]
output = evo_agent.generate("Write a function to compute the factorial of a number.", tools)
print(output)
```
